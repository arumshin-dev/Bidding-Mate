# pyhwp(한글 파일 포맷 분석용 오픈소스):hwp to html

```python
import subprocess
exe_path = r"E:\Miniconda3\envs\p1\Scripts\hwp5html.exe"   # hwp5html.exe 경로
input_file = r"E:\ai\Bidding-Mate\data\processed\test.hwp"                        # 변환할 HWP 파일
output_file = r"E:\ai\Bidding-Mate\team_members\a\test"                      # 생성될 HTML 파일

result = subprocess.run(
    [exe_path, "--output", output_file, input_file],
    capture_output=True,
    text=True
)

print("stdout:", result.stdout)
print("stderr:", result.stderr)
print("returncode:", result.returncode)
```

```python
from bs4 import BeautifulSoup

# HTML 파일 읽기
with open(r"E:\ai\Bidding-Mate\team_members\a\test\index.xhtml", "r", encoding="utf-8") as f:
    html = f.read()

# 파싱
soup = BeautifulSoup(html, "html.parser")

# 본문 텍스트만 추출
text = soup.get_text(separator="\n", strip=True)

print(text)
```

```python
body_text = soup.body.get_text(separator="\n", strip=True)
body_text
```

```python
# 클래스 선택해 읽기
# main_text = soup.find("div", {"class": "main"}).get_text(separator="\n", strip=True)
# main_text

# 스크립트와 스타일 제거
for script in soup(["script", "style"]):
    script.decompose()

clean_text = soup.get_text(separator="\n", strip=True)
clean_text
```

# pyhwp(한글 파일 포맷 분석용 오픈소스):hwp to txt

```python
exe_path = r"E:\Miniconda3\envs\p1\Scripts\hwp5txt.exe"   # hwp5txt.exe 경로
input_file = r"E:\ai\Bidding-Mate\data\processed\test.hwp"                        # 변환할 HWP 파일
output_file = r"E:\ai\Bidding-Mate\team_members\a\test\test.txt"                      # 생성될 HTML 파일

result = subprocess.run(
    [exe_path, "--output", output_file, input_file],
    capture_output=True,
    text=True
)

print("stdout:", result.stdout)
print("stderr:", result.stderr)
print("returncode:", result.returncode)
```
