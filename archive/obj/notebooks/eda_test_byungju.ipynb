```python
import os
import glob
import pymupdf4llm
import olefile
import zlib
import struct
from tqdm import tqdm # ì§„í–‰ìƒí™© ë°”

# ==========================================
# 1. ê²½ë¡œ ì„¤ì • (ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”!)
# ==========================================
# ì˜ˆ: ì›ë³¸ íŒŒì¼ë“¤ì„ 'raw_data'ë¼ëŠ” í´ë”ì— ëª½ë•… ë„£ì–´ë‘ì…¨ë‹¤ë©´ ì•„ë˜ì²˜ëŸ¼ ì“°ì„¸ìš”.
source_dir = "./data/raw/files"   
output_dir = "./final_markdown" 

# ê²°ê³¼ í´ë” ìë™ ìƒì„±
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# ==========================================
# 2. ë³€í™˜ í•¨ìˆ˜ë“¤ (HWP, PDF)
# ==========================================
def load_hwp_as_md(file_path):
    try:
        f = olefile.OleFileIO(file_path)
        dirs = f.listdir()
        if ["BodyText"] not in dirs: return ""
        text_list = []
        sections = [d[1] for d in dirs if d[0] == "BodyText"]
        for section in sections:
            bodytext = f.openstream("BodyText/" + section).read()
            if f.openstream("FileHeader").read(256)[255] == 0:
                 unpacked_data = bodytext
            else:
                try: unpacked_data = zlib.decompress(bodytext, -15)
                except: unpacked_data = bodytext
            
            i = 0
            while i < len(unpacked_data):
                if i + 2 > len(unpacked_data): break
                code = struct.unpack("<H", unpacked_data[i:i+2])[0]
                i += 2
                if code > 31: text_list.append(chr(code))
                elif code == 13: text_list.append("\n")
        
        filename = os.path.basename(file_path)
        return f"# {filename}\n\n" + "".join(text_list)
    except Exception:
        return None # ì‹¤íŒ¨ ì‹œ None ë°˜í™˜

def load_pdf_as_md(file_path):
    try:
        return pymupdf4llm.to_markdown(file_path)
    except Exception:
        return None

# ==========================================
# 3. ë©”ì¸ ì‹¤í–‰ ë¡œì§ (ì—ëŸ¬ ë¡œê·¸ í¬í•¨)
# ==========================================
# ëª¨ë“  íŒŒì¼ ê°€ì ¸ì˜¤ê¸°
all_files = glob.glob(os.path.join(source_dir, "*.*"))
print(f"ì´ {len(all_files)}ê°œì˜ íŒŒì¼ì„ ë³€í™˜í•©ë‹ˆë‹¤.\n")

success_count = 0
failed_files = []

# tqdmìœ¼ë¡œ ì§„í–‰ë¥  ë°” í‘œì‹œ
for file_path in tqdm(all_files, desc="ë³€í™˜ ì¤‘"):
    filename = os.path.basename(file_path)
    name, ext = os.path.splitext(filename)
    ext = ext.lower()
    
    md_content = None
    
    # í™•ì¥ìë³„ ì²˜ë¦¬
    if ext == ".pdf":
        md_content = load_pdf_as_md(file_path)
    elif ext == ".hwp":
        md_content = load_hwp_as_md(file_path)
    else:
        continue # pdf, hwp ì•„ë‹ˆë©´ ìŠ¤í‚µ

    # ì €ì¥
    if md_content:
        save_path = os.path.join(output_dir, name + ".md")
        with open(save_path, "w", encoding="utf-8") as f:
            f.write(md_content)
        success_count += 1
    else:
        failed_files.append(filename)

# ==========================================
# 4. ê²°ê³¼ ë¦¬í¬íŠ¸
# ==========================================
print("\n" + "="*30)
print(f"ì™„ë£Œ! ì„±ê³µ: {success_count}ê°œ")
if failed_files:
    print(f"ì‹¤íŒ¨: {len(failed_files)}ê°œ")
    print("--- ì‹¤íŒ¨í•œ íŒŒì¼ ëª©ë¡ ---")
    for f in failed_files:
        print(f)
else:
    print("ëª¨ë“  íŒŒì¼ì´ ì„±ê³µì ìœ¼ë¡œ ë³€í™˜ë˜ì—ˆìŠµë‹ˆë‹¤!")
print(f"ì €ì¥ ìœ„ì¹˜: {output_dir}")
```

```python
import os
import shutil
import pandas as pd
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
from langchain.docstore.document import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
import re

# ì„¤ì • ë³€ê²½ (2000ì ì‹¤í—˜ìš©)
# ê²¹ì¹˜ëŠ” êµ¬ê°„(Overlap)ë„ 200 -> 400ìœ¼ë¡œ ë¹„ìœ¨ì— ë§ì¶° ëŠ˜ë ¤ì¤ë‹ˆë‹¤.
CHUNK_SIZE = 2000
CHUNK_OVERLAP = 400
NEW_DB_PATH = "./chroma_db_2000"  # ê²½ë¡œë¥¼ ë‹¤ë¥´ê²Œ ì„¤ì •

# ê¸°ì¡´ 2000ì í´ë”ê°€ ìˆë‹¤ë©´ ì‚­ì œ (ì´ˆê¸°í™”)
if os.path.exists(NEW_DB_PATH):
    shutil.rmtree(NEW_DB_PATH)

# ë°ì´í„° ë¡œë“œ ë° ì „ì²˜ë¦¬
def clean_text_final(text):
    if not isinstance(text, str): return ""
    text = text.replace('\r', '')
    text = re.sub(r'[-=]{3,}', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

file_path = "data/raw/data_list.csv"
df = pd.read_csv(file_path)
df_clean = df.dropna(subset=['í…ìŠ¤íŠ¸']).copy()
df_clean['í…ìŠ¤íŠ¸'] = df_clean['í…ìŠ¤íŠ¸'].apply(clean_text_final)

documents = []
for index, row in df_clean.iterrows():
    combined_text = f"[ì‚¬ì—…ëª…: {row['ì‚¬ì—…ëª…']}] \n {row['í…ìŠ¤íŠ¸']}"
    documents.append(Document(
        page_content=combined_text,
        metadata={'source': str(row['ê³µê³  ë²ˆí˜¸']), 'title': row['ì‚¬ì—…ëª…']}
    ))

# ==========================================
# 3. ì²­í‚¹ (ì—¬ê¸°ê°€ í•µì‹¬ ë³€ê²½!)
# ==========================================
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=CHUNK_SIZE,       # 2000ì
    chunk_overlap=CHUNK_OVERLAP, # 400ì
    add_start_index=True
)
splits = text_splitter.split_documents(documents)
print(f"2000ì ë‹¨ìœ„ ë¶„í•  ê²°ê³¼: ì´ {len(splits)}ê°œì˜ ì¡°ê° ìƒì„± (1000ìì¼ ë•Œë³´ë‹¤ ê°œìˆ˜ê°€ ì¤„ì—ˆì„ ê²ë‹ˆë‹¤)")

# ==========================================
# 4. ë²¡í„° DB ìƒì„±
# ==========================================
embedding_model = OpenAIEmbeddings(model="text-embedding-3-small", chunk_size=100)

print(f"2000ì ë²„ì „ DB ìƒì„± ì¤‘... (ê²½ë¡œ: {NEW_DB_PATH})")
vectorstore_2000 = Chroma.from_documents(
    documents=splits,
    embedding=embedding_model,
    collection_name="bidding_rag_db_2000",
    persist_directory=NEW_DB_PATH
)
print("2000ì DB ìƒì„± ì™„ë£Œ!")
```

```python
import pandas as pd
import os
import win32com.client # HWPìš©
import pdfplumber      # PDFìš©

def extract_text_from_hwp(file_path):
    """pywin32ë¥¼ ì´ìš©í•´ í•œê¸€ í”„ë¡œê·¸ë¨ì„ ë°±ê·¸ë¼ìš´ë“œ ì‹¤í–‰ í›„ í…ìŠ¤íŠ¸ ì¶”ì¶œ"""
    try:
        hwp = win32com.client.Dispatch("HWPFrame.HwpObject")
        hwp.XHwpWindows.Item(0).Visible = False # ì°½ ìˆ¨ê¸°ê¸°
        hwp.RegisterModule("FilePathCheckDLL", "FilePathCheckerModule") # ë³´ì•ˆ íŒì—… ìŠ¹ì¸
        hwp.Open(file_path)
        text = hwp.GetTextFile("TEXT", "") # í…ìŠ¤íŠ¸ ëª¨ë“œë¡œ ì¶”ì¶œ
        return text
    except Exception as e:
        return ""
    finally:
        hwp.Clear(1)
        hwp.Quit()

def extract_text_from_pdf(file_path):
    """pdfplumberë¥¼ ì´ìš©í•´ í˜ì´ì§€ë³„ í…ìŠ¤íŠ¸ ë³‘í•©"""
    full_text = ""
    try:
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                full_text += page.extract_text() + "\n"
        return full_text
    except:
        return ""

def load_and_process_data(csv_path):
    """CSV ëª©ë¡ì„ ìˆœíšŒí•˜ë©° í™•ì¥ìì— ë§ëŠ” ì¶”ì¶œê¸° ì‹¤í–‰"""
    df = pd.read_csv(csv_path)
    texts = []
    
    for _, row in df.iterrows():
        path = f"data/files/{row['íŒŒì¼ëª…']}"
        ext = os.path.splitext(path)[1].lower()
        
        if ext == '.hwp':
            texts.append(extract_text_from_hwp(path))
        elif ext == '.pdf':
            texts.append(extract_text_from_pdf(path))
        else:
            texts.append("")
            
    df['í…ìŠ¤íŠ¸'] = texts
    return df
```
